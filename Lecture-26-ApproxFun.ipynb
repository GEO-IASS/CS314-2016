```julia
using Plots
```

```julia
x = Fun(x -> x,[-1,1])
```

```julia
f = 1/(1+x^2)
```

```julia
# f is a polynomial at Chebyshev points.
length(f.coefficients)
```

```julia
g = f+x
plot(g)
```

```julia
plot(g') # this is the derivative!
```

```julia
gppppp = differentiate(g, 5)
plot(gppppp)
```

```julia
maximum(gppppp)
```

```julia
h = sin(g)
z = abs(x)
```

```julia
roots(h + z)
```

```julia
plot(h+z)
```

```julia
roots(h+z-1)
```

```julia
plot(cumsum(h+z-1)) # this is the integral!
```

```julia
sum(h+z-1)
```
